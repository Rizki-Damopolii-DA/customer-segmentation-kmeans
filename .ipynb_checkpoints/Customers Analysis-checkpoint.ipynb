```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
df_customers = pd.read_csv("Mall_Customers.csv")
```

```python
df_customers.head()
```

```python
df_customers = df_customers.rename(columns={"Genre" : "Gender"})
```

```python
df_customers.info()
```

```python
df_customers.describe()
```

```python
gend_pct = df_customers['Gender'].value_counts(normalize=True)*100
gend_pct.round(1).astype(str) + '%'
```

```python
X = df_customers[['Annual Income (k$)',	'Spending Score (1-100)']]
```

```python
X
```

```python
sns.scatterplot(
    data=X,
    x='Annual Income (k$)',
    y='Spending Score (1-100)'
)
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []
sil = []
dbs = []

K = range(2,10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    wcss.append(kmeans.inertia_)
    sil.append(silhouette_score(X_scaled, labels))
    dbs.append(davies_bouldin_score(X_scaled, labels))
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.plot(K, wcss, marker='o')
plt.title('Elbow (WCSS)')
plt.xlabel('k')

plt.subplot(1,3,2)
plt.plot(K, sil, marker='o')
plt.title('Silhouette Score')
plt.xlabel('k')

plt.subplot(1,3,3)
plt.plot(K, dbs, marker='o')
plt.title('Davies Bouldin Index')
plt.xlabel('k')

plt.savefig("../optimal k.png", dpi=300, bbox_inches="tight")

plt.show()
```

##### All three evaluation metrics consistently indicate that k=5 provides the most optimal clustering structure.

```python
MainClusters = KMeans(n_clusters=5, random_state=42)
clusters = MainClusters.fit_predict(X_scaled)
df_customers['Cluster'] = clusters
```

```python
df_customers.head()
```

```python
centroids = scaler.inverse_transform(MainClusters.cluster_centers_)

sns.scatterplot(x=df_customers['Annual Income (k$)'],
                y=df_customers['Spending Score (1-100)'],
                hue=df_customers['Cluster'],
                palette='Set1')

plt.scatter(
    centroids[:,0],
    centroids[:,1],
    s=100,
    c='black',
    marker='X',
    label='Centroid'
)

plt.savefig("../clusters.png", dpi=300, bbox_inches="tight")
plt.legend()
```

#### Cluster 0 represents custommer with moderate income and moderate spending levels, forming the most mainstream customer base of the mall. (Mainstream)  
#### Cluster 1 represents the highest-value segment with both high income and high spending. (Most Valuable Group)  
#### Cluster 2 consists of low-income yet high-spending customers, suggesting impulsive or aspirational buyers.  (Impulsive Buyer)  
#### Cluster 3 shows high-income but low-spending customers, indicating an under-engaged premium segment. (Most Potential Group)  
#### Cluster 4 consist of low-income and low-spending customers, represents the lowest-value segment in terms of revenue contribution. (Low value)

```python
segment_names = {
    0: "Mainstream Customers",
    1: "Affluent Enthusiasts (Most Valuable)",
    2: "Impulsive Spenders",
    3: "Wealthy Conservatives (Most Potential)",
    4: "Budget Customers (Low Value)"
}

df_customers['Segment'] = df_customers['Cluster'].map(segment_names)
```

```python
cluster_summary = df_customers.groupby(['Cluster', 'Segment']).agg({
    'Age':'mean',
    'Annual Income (k$)':'mean',
    'Spending Score (1-100)':'mean',
    'CustomerID':'count'
}).rename(columns={'CustomerID':'Count'})

cluster_summary['Percentage'] = cluster_summary['Count'] / len(df_customers) * 100
```
